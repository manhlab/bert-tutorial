### Задача про блуждание (35 баллов)

Предположим, что в начале движения Шерлок находится в точке с координатой $0$. В каждый момент времени его координата изменяется на $\Delta x$ метров, где $\Delta x$ — случайная величина, имеющая стандартное нормальное распределение. 1. Постройте график изменения положения Шерлока (его координаты на прямой) за $1000$ шагов. По горизонтали – количество пройденных шагов, по вертикали – положение Шерлока.[**5 баллов**]

2. Постройте для траектории положения Шерлока из предыдущего пункта график изменения выборочного среднего и дисперсии для $\Delta x$. По горизонтали – количество шагов, по вертикали – значения выборочного среднего и выборочной дисперсии. [**10 баллов**]

3. Проведите $10^3$ экспериментов по 500 шагов. Продемонстрируйте на одном графике $15$ случайных траекторий из проведенных экспериментов. [**10 баллов**]

4. Добавьте на график из предыдущего пункта усредненную по всем экспериментам траекторию выборочной дисперсии для каждого шага. [**10 баллов**]

```python
posion = 0
arr = []
for i in range(1000):
    x = np.random.normal(0,1)
    posion += x
    arr.append(posion)
plt.plot(arr)
```

```python
stds = []
means = []
arr = np.array(arr)
for i in  range(1000):
    stds.append(arr[:i].std())
    means.append(arr[:i].mean())
plt.plot(stds)
plt.plot(means)
plt.show()
```

```python
summary_posion = []
summary_std = []
for j in range(1000):
    stds= []
    arr = np.array([])
    posion = 0
    for i in range(500):
        x = np.random.normal(0,1)
        posion += x
        arr = np.append(arr, posion)
        stds.append(arr.std())
    summary_std.append(stds)
    summary_posion.append(arr)
#random 15 index of array
ids = np.random.randint(1,1000,15)
for id in ids:
    plt.plot(summary_posion[id])
    plt.plot(summary_std[id])
plt.show()
```
